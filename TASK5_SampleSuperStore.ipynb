### Sub-Category vs Profit

```python
fig,ax=plt.subplots(figsize=(10,6))
x=df['Sub-Category']
y=df['Profit']
ax.set_xlabel('Products')
ax.set_ylabel('Profit Earned')
ax.set_title('Bar')

  
fig.autofmt_xdate()

ax.bar(x,y)
plt.show()
```

#### Conclusion: Copiers have Most Profit and Machine has least profit

### In which Segment there is more sales?

```python
sns.countplot(df["Segment"]);
```

#### Conclusion: The most sales is in Consumer Segment & least sales is in Home Office Segment

### In which Region there is more sales?

```python
sns.countplot(df["Region"]);
```

####  Conclusion: West has most sales and South has least sales

### State vs No of Sales

```python
fig,ax=plt.subplots(figsize=(20,10))
sns.countplot(df["State"],order = df['State'].value_counts().index);
fig.autofmt_xdate()
```

#### Conclusion: California has most items sold and Wyoming has least items sold.

### Sales & Profit Based on Ship Mode

```python
df2=df.groupby("Ship Mode").sum()[["Profit","Sales"]]
df2.plot.bar();
```

#### Conclusion: Standard Class Shipping Mode has most Profit & Sales and  Same Day Shipping Mode has Least Profit & Sales

### In which Region there is more Profit?

```python
sns.barplot(df["Region"],df["Profit"],ci=None);
```

#### Conclusion: There in more Profit in West and less profit in Central region

### Category Vs Sales, Impact of Different Categories on Sales

```python
sns.barplot(df["Category"],df["Sales"],ci=None);
```

#### Conclusion: Technology has most sales than Furniture and Office Supplies Category

### Subcategory Counts

```python
fig,ax=plt.subplots(figsize=(11,11))
df["Sub-Category"].value_counts().plot.pie(autopct='%.2f',shadow=True);
```

#### Conclusion: Blinders are sold most and Copiers are sold least

```python
sns.catplot("Ship Mode",data=df,hue="Segment",kind="count");
```

#### Conclusion: Most people prefer Standard Class Shipping Mode in every category whether it is Consumer, Corporate or Home Office and same day Shipping mode is less preffered may be because of additional Shipping Expenses of Same day Shipping Mode.

### Impact of different Sub - Categories on Sales

```python
fig,ax=plt.subplots(figsize=(10,6))
chart=sns.barplot(df["Sub-Category"],df["Sales"],ci=None)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

#### Conclusion: Copiers has most sales and Fasteners has least sales.

### Impact of Postal Codes, Sales, Quantity, Discount and Profit under various Sub-Categories

```python
sns.pairplot(df,hue="Sub-Category");
```

### State vs Catgeory Vs Profit

```python
plt.figure(figsize=(10,12))
dfheat1=pd.pivot_table(df,values="Profit",index=["State"],columns="Category")
sns.heatmap(dfheat1,cmap="YlGnBu");
```

#### Conclusion : We can find that Technical items have more profit in various states especially in District of Columbia and Rhode Island. But there is least profit in Office Supplies Category in almost all states.

```python
plt.figure(figsize=(10,12))
dfheat1=pd.pivot_table(df,values="Profit",index=["State"],columns="Sub-Category")
sns.heatmap(dfheat1,cmap="YlGnBu");
```

#### Conclusion: We have already seen Copiers has most sales and they are mostly in Delaware and Indiana

### Sub-Category Vs Profit Vs Discount

```python
plt.figure(figsize=(10,12))
dfheat2=pd.pivot_table(df,values="Profit",index=["Sub-Category"],columns="Discount")
sns.heatmap(dfheat2,cmap="YlGnBu");
```

#### Conclusion : We can see that as Discount rates increases there is a decline in profits earned in almost every sub-categories. Therefore reducing the discount rates can earn more profits
